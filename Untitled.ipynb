В предыдущем шаге, были выявлены проблемы с названиями, было принято решение испавить их используя команду rename

```python
df = df.rename(columns={"Name": "name","Platform":"platform","Year_of_Release":"year_of_release",
                        "Genre":"genre","NA_sales":"na_sales","EU_sales":"eu_sales",
               "JP_sales":"jp_sales","Other_sales":"other_sales","Critic_Score":"critic_score",
               "User_Score":"user_score","Rating":"rating"})
```

```python
df
```

Поскольку в работе были обнаружены пропуски, требуется применить команду fillna(0),чтобы не лишиться строк с важными данными, заполним значения NaN

```python
df['name'] = df['name'].fillna(0)
df['platform'] = df['platform'].fillna(0)
df['year_of_release'] = df['year_of_release'].fillna(0)
df['genre'] = df['genre'].fillna(0)
df['na_sales'] = df['na_sales'].fillna(0)
df['eu_sales'] = df['eu_sales'].fillna(0)
df['jp_sales'] = df['jp_sales'].fillna(0)
df['other_sales'] = df['other_sales'].fillna(0)
df['critic_score'] = df['critic_score'].fillna(0)
df['user_score'] = df['user_score'].fillna(0)
df['rating'] = df['rating'].fillna(0)
```

проверка данных на наличие дубликатов, если они есть – произойдет удаление

```python
df = df.drop_duplicates()
df = df.drop_duplicates().reset_index()
df
```

```python
def convert_currency(val):
    """
    Преобразует числовое значение строки в число с плавающей точкой:
     - удаляет $
     - удаляет запятые
     - преобразует в число с плавающей точкой
    """
    new_val = val.replace(',', '')
    return float(new_val)
```
